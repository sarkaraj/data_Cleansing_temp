```python
# Python
import pandas as pd
import numpy as np
from fbprophet import Prophet
from dateutil import parser
import matplotlib.pylab as plt
%matplotlib inline
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 15, 6
```

```python
# import os

# os.chdir('C:\Users\dev2')

# %pwd
```

```python
data = pd.read_csv('prod1_deli_mart_monthly.csv', dtype={'Date' : str, 'Quantity' : float})
# data.Date = data.Date.apply(str)
data.Date = data.Date.apply(parser.parse)
data = data.rename(columns={'Date': 'ds', 'Quantity': 'y'})
data = data[['ds','y']]

data.tail()
```

```python
plt.plot(data.ds,data.y)
```

```python
m = Prophet()
m.fit(data);
```

```python
future = m.make_future_dataframe(periods=0, freq= 'M')
future.tail()
```

```python
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
ax = data.y.plot(label='Observed')
forecast.yhat.plot(ax=ax, label='Forecast', alpha=.7)

ax.fill_between(forecast.index,
                forecast.yhat_lower,
                forecast.yhat_upper, color='k', alpha=.1)

ax.set_xlabel('Index')
ax.set_ylabel('Target')
plt.legend()

plt.show()
```

```python
y_forecasted = forecast.yhat
y_truth = data.y

# Compute the mean square error
mse = (((y_forecasted - y_truth) ** 2).mean())**0.5
mape = np.mean(np.abs((y_truth - y_forecasted) / y_truth)) * 100
print('The Mean Squared Error of our forecasts is {}'.format(round(mse, 2)))
print('The Mean Absolute percentage Error of our forecasts is {}'.format(round(mape, 2)))
```
